**The grades have been collected over a number of years, so we should merge the grade columns**

```python
grades_cols=['Grades1997','Grades1998','Grades1999','Grades2000','Grades2001','Grades2002','Grades2003','Grades2004','Grades2005','Grades2006','Grades2007']
```

```python
for year in grades_cols:
    print(df[year].value_counts())
```

**There's lots of special values (like -4 which indicates that there is no information for this particular year) so we should deal with them individually.
Grades ranging from 1 to 8 exist on an ordinal scale so we should leave them as they are.**

```python
# taking care of -4
for year in grades_cols:
    df[year]=np.where(df[year]==-4,0,df[year])
```

```python
for year in grades_cols:
    print(df[year].value_counts())
```

```python
# taking care of -5
for year in grades_cols:
    df[year]=np.where(df[year]==-5,0,df[year])
```

```python
for year in grades_cols:
    print(df[year].value_counts())
```

**9   Other (SPECIFY) <br/>
10  As to Cs <br/>
11  Mixed <br/>
12 Ungraded <br/>
will be replaced with the mode**

**-1 Refusal <br/>
-2 Don't know <br/>
will be removed, since they constitute a small number of participants**

```python
for year in grades_cols:
    mode=df[year].value_counts()[1:].index[0]
    df[year]=np.where(df[year]==9,mode,df[year])
    df[year]=np.where(df[year]==10,mode,df[year])
    df[year]=np.where(df[year]==11,mode,df[year])
    df[year]=np.where(df[year]==12,mode,df[year])
```

```python
for year in grades_cols:
    print(df[year].value_counts())
```

```python
# taking care of -1 and -2
for year in grades_cols:
    df = df.drop(df[df[year]==-1].index)
    df = df.drop(df[df[year]==-2].index)
```

```python
for year in grades_cols:
    print(df[year].value_counts())
```

```python
# dealing with repeating grades
repeating_indices=[]
indices_for_dropping=[]
for index, row in df.iterrows():
    if len(row[grades_cols].value_counts())==2:
        for ind, val in  row[grades_cols].value_counts()[1:].items():
            if(val!=1):
                repeating_indices.append(index)
    else:
        indices_for_dropping.append(index)
```

```python
print(len(indices_for_dropping))
print(indices_for_dropping)
```

```python
print(len(df))
df=df.drop(indices_for_dropping)
print(len(df))
```

```python
print(len(repeating_indices))
print(repeating_indices)
```

```python
df['HighSchoolGrade']=df.Grades1997+df.Grades1998+df.Grades1999+df.Grades2000+df.Grades2001+df.Grades2002+df.Grades2003+df.Grades2004+df.Grades2005+df.Grades2006+df.Grades2007
```

```python
df.loc[repeating_indices].HighSchoolGrade=df.loc[repeating_indices].HighSchoolGrade/2
```
